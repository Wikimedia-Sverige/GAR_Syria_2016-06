```python
all_tokens = []
for index, string in metadata["Place"].iteritems():
    tokens = string.split()
    all_tokens.extend(tokens)
all_tokens
```

```python
cnt = Counter()
for token in all_tokens:
    cnt[token] += 1
cnt.most_common(30)
```

```python
count = pd.DataFrame(cnt.most_common(200))
count.columns = ["name","frequency"]
count["category"] = None
count.set_index("name")
count.to_csv("./Place_keyword_freqs.csv")
```
